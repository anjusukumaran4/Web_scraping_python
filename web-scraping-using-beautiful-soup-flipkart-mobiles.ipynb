```python
#Packages
#--Web scraping packages
from bs4 import BeautifulSoup
import requests
#Pandas/numpy for data manipulation
import numpy as np
import pandas as pd
```

```python
#scrap details about iphones from flipkart
number_of_pg =1

#creating empty list
title=[]
price=[]
memory=[]
camera=[]
rating=[]
```

```python

for i in np.arange(0,number_of_pg):
    base_url=requests.get('https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off')
    cover_page=base_url.text
    soup=BeautifulSoup(cover_page,'html.parser')
    title_mbl=soup.find_all('div',class_='_4rR01T')
    price_mbl=soup.find_all('div',class_='_25b18c')
    memory_mbl=soup.find_all('li',class_='rgWa7D')
    camera_mbl=soup.find_all('li',class_='rgWa7D')
    rating_mbl=soup.find_all('span',class_='_2_R_DZ')
    for t,p,m,c,r in zip(title_mbl,price_mbl,memory_mbl,camera_mbl,rating_mbl):
        title.append(t.text)
        price.append(p.text)
        memory.append(m.text)
        camera.append(c.text)
        rating.append(r.text)
```

```python
print(title)
```

**DataFrame**

```python
df=pd.DataFrame({'Title':title,'Price':price,'Memory':memory,'Camera':camera,'Rating':rating})
```
